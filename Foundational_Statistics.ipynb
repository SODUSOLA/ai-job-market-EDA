### Dataset: AI Job Market Dataset

##### Imports

```python
import pandas as pd
```

__Load the dataset__

```python
# Load the dataset
data = pd.read_csv('ai_job_dataset.csv')

print("Dataset loaded successfully\n")
print(f"Shape of the dataset: {data.shape}")
```

__Exploring the data__

```python
# Display the first 5 rows
print("\n--- First 5 rows of the dataset ---")
display(data.head())
```

```python
# Check data types and structure
print("\n--- Dataset Information ---\n")
data.info()
```

```python
# Display summary statistics for numerical columns

print("\n--- Summary Statistics (Numerical Columns ---)\n")
display(data.describe())
```

__Basic Statistical Analysis__

```python
# Selecting numerical columns only

numerical_data = data.select_dtypes(include='number')

# Stats (mean, median, and standard deviation)

stats = numerical_data.agg(['mean', 'median', 'std'])
print("\n--- Mean, Median and Standard Deviation---\n")
display(stats)
```

```python
# Mode of numerical features

mode = numerical_data.mode().iloc[0]
print("\n--- Mode of Numerical Features ---\n")
print(mode)
```

__Correlation Analysis__

```python
# Correlation matrix for numerical features

correlation_matrix = numerical_data.corr()
print("\n--- Correlation Matrix ---\n")
display(correlation_matrix)
```
